### Seasonality

#### Decomposition

- Trend -> pattern;
- Seasonal -> cyclical events
- Residual -> error of prediction.

#### Naive Decomposition (one of the simplest)

- Additive: observed = trend + seasonal + residual 
- Multiplicative: observed = trend x seasonal x residual

```python
import pandas as pd
import numpy as np
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from pandas.plotting import autocorrelation_plot
sns.set()
```

```python
rw = pd.read_csv("df_2018_with_rw.csv")
df = rw.copy()
df["date"] = pd.to_datetime(df["date"], dayfirst=True) 
df.set_index("date", inplace=True)
df = df.asfreq("b")
df = df.fillna(method="ffill")
#df = df.dropna()
```
